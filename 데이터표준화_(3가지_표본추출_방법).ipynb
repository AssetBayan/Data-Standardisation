```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset("tips")

# просмотр первых строк
tips.head()
```

```python
sample_random = tips.sample(n=20, random_state=42)

sample_random
```

```python
step = 10
sample_systematic = tips.iloc[::step]

sample_systematic
```

```python
sample_stratified = tips.groupby("sex").sample(n=10, random_state=42)

sample_stratified
```
